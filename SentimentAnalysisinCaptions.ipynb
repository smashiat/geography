```python
import os
import random
import numpy as np
from pathlib import Path
from google.colab import drive
from collections import Counter
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Create a DataFrame
data = {'Category': ['Africa', 'Europe', 'Americas', 'East Asia','South East Asia', 'West Asia' ],
        'NumofImages': [0, 0, 0, 0, 0, 0], 'Mean': [0, 0, 0, 0, 0, 0], 'Std Dev' : [0, 0, 0, 0, 0, 0] }

df = pd.DataFrame(data)

# Save to CSV in your Drive
df.to_csv('/content/drive/My Drive/negative.csv', index=False)


drive.mount('/content/drive')
# get the path/directory
file_path = ["/content/drive/MyDrive/Colab Notebooks/captions/africa_bicycle.txt",
             "/content/drive/MyDrive/Colab Notebooks/captions/africa_car.txt",
             "/content/drive/MyDrive/Colab Notebooks/captions/africa_truck.txt",
             "/content/drive/MyDrive/Colab Notebooks/captions/africa_bus.txt",
             "/content/drive/MyDrive/Colab Notebooks/captions/africa_boat.txt"]

output_path = "/content/drive/MyDrive/Colab Notebooks/stats/africa_negative.txt"
# Ensure the folder exists
target_labels = ["dirty", "messy", "wrecked", "vandaled", "damaged", "broken", "crash", "crushed", "crashed", "accident"]
count = 0
current_count = []

for fp in file_path:
  with open(fp, 'r') as file:
    lines = file.readlines()
    for i in range(0, len(lines), 2):
        filename = lines[i].strip()
        caption = lines[i+1].strip()
        count += 1
        if any(label in caption for label in target_labels):
          current_count.append(1)
        else:
          current_count.append(0)

# Update cells
df.loc[df['Category'] == 'Africa', 'NumofImages'] = count

label_counts = []
means = []
converged = False
iterations = 0
mean = 0
std_dev = 0
total = 0
with open(output_path, 'w') as output_file:
  for i in range(0, 500):
      # Sample with replacement
      bootstrap_sample = random.choices(current_count, k=count)

      total = sum(bootstrap_sample)
      print(f"iterations", iterations, file=output_file)
      print(f"iterations", iterations)
      print(f"sample", bootstrap_sample, file=output_file)
      print(f"sample", bootstrap_sample)
      print(f"length", len(bootstrap_sample))
      print(f"total", total, file=output_file)
      print(f"total", total)
      new_prob = total / count
      print(f"count", count)
      print(f"new_prob", new_prob)
      label_counts.append(new_prob)
      print(f"label count", label_counts, file=output_file)
      mean = np.mean(label_counts)
      means.append(mean)
      print(f"means", means, file=output_file)
      print(f"means", means)
      std_dev = np.std(label_counts)
      print(f"std dev", std_dev, file=output_file)
      print(f"std dev", std_dev)
      iterations += 1


df.loc[df['Category'] == 'Africa', 'Mean'] = mean
df.loc[df['Category'] == 'Africa', 'Std Dev'] = std_dev
df.to_csv('/content/drive/My Drive/negative.csv', index=False)
```
