```python
import os
import random
import numpy as np
from PIL import Image
from pathlib import Path
from google.colab import drive
from collections import Counter
from transformers import AutoProcessor, BlipForQuestionAnswering


drive.mount('/content/drive')
model = BlipForQuestionAnswering.from_pretrained("Salesforce/blip-vqa-base")
processor = AutoProcessor.from_pretrained("Salesforce/blip-vqa-base")

# get the path/directory
folder_dir = "/content/drive/MyDrive/Colab Notebooks/africa/bicycle"
output_path = "/content/drive/MyDrive/Colab Notebooks/vqa/africa_bicycle_vehicle.txt"
# Ensure the folder exists
image_paths = []
count = 0;
with open(output_path, 'w') as output_file:
  for filename in os.listdir(folder_dir):
      image_path = os.path.join(folder_dir, filename)
      # caption generation code here
      image = Image.open(image_path)
      text = "Is there a vehicle in the picture?"
      print(f"Filename:", image_path , file=output_file)
      inputs = processor(images=image, text=text, return_tensors="pt")
      output = model.generate(**inputs)
      answer = processor.decode(output[0], skip_special_tokens=True)
      print(f"", answer, file=output_file)
```
